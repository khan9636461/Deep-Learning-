```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
img = image.load_img('/content/cat.10.jpg', target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='constant'

)
```

```python
img.shape
```

```python
img = image.img_to_array(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
import os

# Create the 'preview' directory if it doesn't exist
if not os.path.exists('preview'):
    os.makedirs('preview')

# Now run the data augmentation and save images
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='preview', save_format='jpg'):
    i += 1
    if i == 10:
        break
```
