```python
from keras.applications.vgg16 import VGG16

# Load the VGG16 model
model = VGG16()
```

```python
# Print the model summary
model.summary()
```

```python
from keras.utils import plot_model

# Plot the model
plot_model(model)
```

```python
for i in range(len(model.layers)):
	# check for convolutional layer
	if 'conv' not in model.layers[i].name:
		continue
	# get filter weights
	filters, biases = model.layers[i].get_weights()
	print("layer number",i,model.layers[i].name, filters.shape)
```

```python
# retrieve weights  from the secnd hidden layer
filters, biases = model.layers[1].get_weights()
```

```python
print(filters.shape)
```

```python
# Normalize filters values
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
import matplotlib.pyplot as plt

n_filters =6
ix=1
fig = plt.figure(figsize=(15,10))
for i in range(n_filters):
    # get the filters
    f = filters[:,:,:,i]
    for j in range(3):
        # subplot for 6 filters and 3 channels
        plt.subplot(n_filters,3,ix)
        plt.imshow(f[:,:,j] ,cmap='gray')
        ix+=1
#plot the filters
plt.show()
```

```python
from keras.models import Model
model = Model(inputs=model.inputs, outputs=model.layers[i].output)
```

```python
from keras.preprocessing.image import load_img, img_to_array
from numpy import expand_dims
from keras.applications.vgg16 import preprocess_input

image = load_img("/content/Fakhar_Khan.jpg" , target_size=(224,224))

# convert the image to an array
image = img_to_array(image)
# expand dimensions so that it represents a single 'sample'
image = expand_dims(image, axis=0)

image = preprocess_input(image)
```

```python
import matplotlib.pyplot as plt

features = model.predict(image)

num_feature_maps = min(features.shape[3], 64)  # Limit to 64 feature maps

fig = plt.figure(figsize=(20, 15))

for i in range(1, num_feature_maps + 1):  # Ensure index doesn't exceed 64
    plt.subplot(8, 8, i)
    plt.imshow(features[0, :, :, i - 1], cmap='gray')
    plt.axis('off')  # Hide axis for better visualization

plt.show()
```

```python
model2 = VGG16()
```

```python
layer_index = [2,5,9,13,17]

outputs = [model2.layers[i].output for i in layer_index]
model3 = Model(inputs=model2.inputs, outputs=outputs)
```

```python
import matplotlib.pyplot as plt

feature_map = model3.predict(image)

for layer_idx, fmap in zip(layer_index, feature_map):  # Avoid reusing 'i'
    fig = plt.figure(figsize=(20, 15))
    fig.suptitle(f"Layer_{layer_idx}", fontsize=20)

    num_feature_maps = min(fmap.shape[3], 64)  # Prevent exceeding 64 subplots

    for j in range(1, num_feature_maps + 1):  # Use 'j' instead of 'i'
        plt.subplot(8, 8, j)
        plt.imshow(fmap[0, :, :, j - 1], cmap='gray')
        plt.axis('off')  # Hide axis for better visualization

    plt.show()
```
